- Calculating state-wise
    - total popluation
    - total popluation with different religions


```python
# so inorder to get data in group wise like state as given in problem
# we will use groupby method

df.groupby('State_name').Population.sum()
```

```python
# printing population count state wise in descending order


df.groupby('State_name').Population.sum().sort_values(ascending = False)
```

```python
df.groupby('State_name')['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains'].sum()
```

```python
df.columns
```

```python
df.groupby('State_name')['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains'].sum().sort_values(ascending = False, by = 'Christians')
```

- Calculating Male workers in Bihar

```python
male_worker = df[df.State_name == 'BIHAR']['Male_Workers'].sum()
female_worker = df[df.State_name == 'BIHAR']['Female_Workers'].sum()

print("Male Workers in Bihar: ", male_worker)
print("Female Workers in Bihar: ", female_worker)
```
